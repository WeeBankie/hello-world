## Figure and Axis

An example 

```python
fig = plt.figure(figsize=(6,4), facecolor='orange')

rect = [0, 0, 1, 1]
rect1 = [0.1, 0.1, 0.8, 0.8]
rect2 = [0.2,0.2,0.6,0.6]

ax = fig.add_axes(rect,facecolor='lightblue')
ax.set_xlim(0,10)
ax.set_xlim(0,8)
ax.set_xticks([0,2,4,6,8,10])
ax.set_yticks([0,2,4,6,8])
ax.set_title('Embedded Axis Object ax', fontsize=14)

ax1 = fig.add_axes(rect1, facecolor='lightyellow')
ax1.set_xlim(-5,5)
ax1.set_ylim(-2,2)
ax1.set_yticks([-2,-1,0,1,2])
ax1.set_title('Inner Axis ax1', fontsize=12)

ax2 = fig.add_axes(rect2, facecolor='white')
ax2.set_xlim(-np.pi,np.pi)
ax2.set_ylim(-1,1)
ax2.set_xticks([-np.pi,0,np.pi])
ax2.set_yticks([-2,-1,0,1,2])
ax2.grid()
ax2.annotate('An annotation at (0,0)',xy=(0.0,0.0))
ax2.annotate('An annotation at (-1,-1)',xy=(-1.0,-1.0))
ax2.set_title('Innermost Axis ax2', fontsize=11)

x = np.linspace(-np.pi,np.pi,500)
y = np.cos(4*x)

ax2.plot(x,y,lw=1.5,label='$cos(4x)$')
ax2.legend(loc='best')

plt.show()

# print(fig,ax,ax1,ax2)
```

## Axis Ticks, Tick Labels, and Grids

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.sin(x) * np.exp(-x**2/20)

fig, axes = plt.subplots(1, 4, figsize=(12, 3), facecolor = 'lightblue')

axes[0].plot(x, y, lw=2)
axes[0].set_title("default ticks")
axes[1].plot(x, y, lw=2)
axes[1].set_title("set_xticks")
axes[1].set_yticks([-1, 0, 1])
axes[1].set_xticks([-5, 0, 5])

axes[2].plot(x, y, lw=2)
axes[2].set_title("set_major_locator")
axes[2].xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
axes[2].yaxis.set_major_locator(mpl.ticker.FixedLocator([-1, 0, 1]))
axes[2].xaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))
axes[2].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(8))

axes[3].plot(x, y, lw=2)
axes[3].set_title("set_xticklabels")
axes[3].set_xlabel(r'angle $\theta$', color='r',fontsize=14)
axes[3].set_yticks([-1, 0, 1])
axes[3].set_xticks([-2 * np.pi, -np.pi, 0, np.pi, 2 * np.pi])
axes[3].set_xticklabels([r'$-2\pi$', r'$-\pi$', 0, r'$\pi$', r'$ 2\pi$'])

x_minor_ticker = mpl.ticker.FixedLocator([-3 * np.pi / 2, -np.pi / 2, 0, np.pi / 2, 3 * np.pi / 2])

axes[3].xaxis.set_minor_locator(x_minor_ticker)
axes[3].yaxis.set_minor_locator(mpl.ticker.MaxNLocator(4))

for label in axes[3].get_xticklabels():
    label.set_color('red')
    # label.set_fontweight('bold') # doesn't work with the LaTeX sysmbols.
    label.set_fontsize(14)
```

**Note the use of np.pi in the x-axis labels. This will be good to set the effective radius, $R_e$.**

## Spines

```python
x = np.linspace(-10, 10, 500)
y = np.sin(x) / x

fig, ax = plt.subplots(figsize=(8, 4), facecolor='lightblue')

ax.plot(x, y, linewidth=2)

# remove top and right spines
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# remove top and right spine ticks
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# move bottom and left spine to x = 0 and y = 0
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

ax.set_xticks([-10, -5, 5, 10])
ax.set_yticks([0.5, 1])

# give each label a solid background of white, to not overlap with the plot line
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_bbox({'facecolor': 'white', 'edgecolor': 'white'})
```

## Tick Labels in Scientific Notation

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

x = np.linspace(0, 1e2, 100)
y = x ** 2

# plt.title('Tick Label Formats',loc='center')

axes[0].plot(x, y, 'r')
axes[0].set_title("default labels", loc='right')

axes[1].plot(x, y, 'b')
axes[1].set_title("scientific notation labels", loc='right')

formatter = mpl.ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,32)) # Use scientific notation for numbers < and > (a,b)
axes[1].xaxis.set_major_formatter(formatter)
axes[1].yaxis.set_major_formatter(formatter)
```

## Twin Axes

```python
fig, ax1 = plt.subplots(figsize=(6, 4))

r = np.linspace(0, 5, 100)
a = 4 * np.pi * r ** 2 # area
v = (4 * np.pi / 3) * r ** 3 # volume

ax1.set_title("surface area and volume of a sphere", fontsize=12)
ax1.set_xlabel("radius [m]", fontsize=12)

ax1.plot(r, a, lw=2, color="blue")
ax1.set_ylabel(r"surface area ($m^2$)", fontsize=12, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(r, v, lw=2, color="red")
ax2.set_ylabel(r"volume ($m^3$)", fontsize=12, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

## Insets

```python
fig = plt.figure(figsize=(8, 4))

def f(x):
    return 1/(1 + x**2) + 0.1/(1 + ((3 - x)/0.1)**2)

def plot_and_format_axes(ax, x, f, fontsize):
    ax.plot(x, f, linewidth=2)
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
    ax.set_xlabel(r"$x$", fontsize=fontsize)
    ax.set_ylabel(r"$f(x)$", fontsize=fontsize)

# main graph
ax = fig.add_axes([0.1, 0.15, 0.8, 0.8], facecolor="#f5f5f5")
x = np.linspace(-4, 14, 1000)
plot_and_format_axes(ax, x, f(x), 18)

# inset
x0, x1 = 2.5, 3.5
ax.axvline(x0, ymax=0.3, color="grey", linestyle=":")
ax.axvline(x1, ymax=0.3, color="grey", linestyle=":")

ax_insert = fig.add_axes([0.5, 0.5, 0.38, 0.42], facecolor='none')
x = np.linspace(x0, x1, 1000)
plot_and_format_axes(ax_insert, x, f(x), 14)
```

## Colour Maps and Colour Bars

Modified code to position the colour bar above the figure.

```python
x = y = np.linspace(-10, 10, 150)
X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.cos(Y) * np.exp(-(X/5)**2-(Y/5)**2)

fig, ax = plt.subplots(figsize=(5, 4))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm=norm, cmap=mpl.cm.bwr)

ax.axis('tight')
ax.set_xlabel(r"$x$", fontsize=14)
ax.set_ylabel(r"$y$", fontsize=14)
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cbaxes = fig.add_axes([0.125, 0.95, 0.775, 0.04]) 
# cb = plt.colorbar(ax1, cax = cbaxes) 

cb = fig.colorbar(p, ax=ax, cax=cbaxes, orientation="horizontal")
# cb.set_label(r"$z$", fontsize= 18)
cb.set_ticks([-1.0, -.5, 0, .5, 1.0])
# cb.set_ticks([])

plt.title('Colour Map Plot')
plt.show()
```

A useful colorbar function:

See: https://joseph-long.com/writing/colorbars/

```python
def colorbar(mappable,loc='right'):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    ax = mappable.axes
    fig = ax.figure
    divider = make_axes_locatable(ax)
    cax = divider.append_axes(loc, size="4%", pad=0.1)
    return fig.colorbar(mappable, cax=cax)
```

## GridSpec layouts

```python
fig = plt.figure(figsize=(14, 10))
gs = mpl.gridspec.GridSpec(4, 4, wspace=0.20, hspace=0.40)
ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[1, 1])
ax2 = fig.add_subplot(gs[2, 2])
ax3 = fig.add_subplot(gs[3, 3])
ax4 = fig.add_subplot(gs[0, 1:])
ax5 = fig.add_subplot(gs[1:, 0])
ax6 = fig.add_subplot(gs[1, 2:])
ax7 = fig.add_subplot(gs[2:, 1])
ax8 = fig.add_subplot(gs[2, 3])
ax9 = fig.add_subplot(gs[3, 2])

ax0.set_title("Uncorrelated")
ax0.scatter(x1, x2)

ax1.set_title("Weakly positively correlated")
ax1.scatter(x1, x1 + x2)

ax2.set_title("Weakly negatively correlated")
ax2.scatter(x1, -x1 + x2)

ax3.set_title("Strongly correlated")
ax3.scatter(x1, x1 + 0.15 * x2)
```

### Another GridSpec example...

```python
fig = plt.figure(figsize=(12, 8))
gs = mpl.gridspec.GridSpec(2, 3,width_ratios=[2, 1.5, 1.5],height_ratios=[1, 2],wspace=0.25, hspace=0.25)
ax0 = fig.add_subplot(gs[1, 0])
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

fig.suptitle('A figure containing a grid array of axes objects', fontsize=18)

x = np.linspace(0, 10, num=100)
y = x ** 2 + 10 * np.random.randn(100)

scatters = ax0.scatter(x, y, c=y, cmap='magma')
colorbar(scatters)
ax0.set_title('Scatter plot with colour bar', fontsize=14)

sine = ax1.plot(x,np.sin(x),lw=1.5,color='b')
ax1.set_title(r'$sin(x)$', fontsize=14)
ax1.set_xlim(0,2*np.pi)
ax1.set_xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi])
ax1.set_xticklabels([0, r'$\pi/2$', r'$\pi$', r'$3\pi/2$',r'$2\pi$'])
ax1.set_yticks([-1,-0.5,0,0.5,1])
ax1.set_yticklabels(['a','b','c','d','e'])

tan = ax2.plot(x,np.tan(x),lw=1.5,color='r')
ax2.set_xlim(0,2*np.pi)
ax2.set_ylim(-15,15)
ax2.set_yticks([-15,-10,-5,0,5,10,15])
ax2.set_xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi])
ax2.set_xticklabels([0, r'$\pi/2$', r'$\pi$', r'$3\pi/2$',r'$2\pi$'])
ax2.set_title(r'$tan(x)$', fontsize=14)

ax3 = fig.add_subplot(gs[0,2])
ax4 = fig.add_subplot(gs[1,1])
ax5 = fig.add_subplot(gs[1,2])

x1 = np.random.randn(100)
x2 = np.random.randn(100)

ax4.set_title("Weakly negatively correlated", fontsize=14)
ax4.scatter(x1, -x1 + x2)

ax3.scatter(x1,x2)
ax3.set_title('uncorrelated', fontsize=14)

x = np.linspace(-10, 10, 100)
y = np.sin(x) / x
ax5.plot(x, y, lw=2, color='g')
ax5.set_title(r'$sin(x)/x$', fontsize=14)
# remove top and right spines
ax5.spines['top'].set_color('none')
ax5.spines['right'].set_color('none')
# move bottom and left spine to x = 0 and y = 0
ax5.spines['bottom'].set_position(('data', 0))
ax5.spines['left'].set_position(('data', 0))
ax5.set_xticks([-10,-5,5,10])
ax5.set_yticks([0.5,1])

fig.savefig('gridtest.png')

plt.show()
# display(fig)
```
