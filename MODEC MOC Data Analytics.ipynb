# Jupyter Notebook to plot MODEC MOC Statistics
Create a proffesionally styled barchart plot grouped by Project.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
sns.set(style='darkgrid')
sns.set_palette('bright') ### 'pastel' or 'bright' palettes work well.
```

### Create a DataFrame of the Data Analytics values

```python
df = pd.DataFrame([['MV32', 0, 0, 147, 116, 0, 142, 0], ['MV33', 25, 2, 120, 238, 0, 144, 14], ['Sangomar', 13, 0, 25, 20, 8, 120, 119]],
                  columns=['Project', 'EA - Engineering Amendment', 'DDN - Design Deviation Notice', 'ECN - Engineering Change Notice',
                           'SI - Site Instruction', 'SQ - Site Query', 'TQ - Technical Query', 'VOR - Variation Order Request'])
```

### View the DataFrame

```python
df.style.hide_index()
```

### Generate the plot

```python
fig, ax1 = plt.subplots(1, 1, figsize=(12,8), facecolor='lightblue')

# plot the grouped bar chart for the DataFrame...
df.plot(x='Project',
        kind='bar',
        width=0.80,
        stacked=False,
        rot=0,
        fontsize=18,
        ax=ax1)

# set the label position offsets...
hoff = 0.0
voff = 3.0

for p in ax1.patches:
    ax1.annotate(str(p.get_height()).center(4), (p.get_x() + hoff, p.get_height() + voff), fontsize=14)
    
# annotate the chart title, axis labels and legend... 
ax1.set_title('Engineering Change Management Documents by Project', fontsize=20)
ax1.set_ylabel('Number of Engineering MOC Documents', fontsize=18)
ax1.set_xlabel('Project', fontsize=20)
ax1.legend(loc='best', fontsize=14)
ax1.set_ylim(0,275)

# save the figure as a .png image...
plt.savefig('MOCstats.png', facecolor='lightblue')
```
