```python
# create data
df = pd.DataFrame([['MV32', 0, 0, 147, 116, 0, 142, 0], ['MV33', 25, 2, 120, 238, 0, 144, 14], ['Sangomar', 13, 0, 25, 20, 8, 120, 119]],
                  columns=['Project', 'EA - Engineering Amendment', 'DDN - Design Deviation Notice', 'ECN - Engineering Change Notice', 'SI - Site Instruction', 'SQ - Site Query', 'TQ - Technical Query', 'VOR - Variation Order Request'])
# view data
print(df)

hoff = 0.0
voff = 3.0

# fig, ax = plt.subplots()
# plt.figure(figsize=(16,12))
   
# plot the grouped bar chart
ax = df.plot(x='Project',
        kind='bar',
        width=0.70,
        stacked=False,
        rot = 1,
        # title='Engineering Change Management documents by Project', 
        fontsize=15,
        figsize=(12,8)
    )

for p in ax.patches:
    ax.annotate(str(p.get_height()).center(3), (p.get_x() + hoff, p.get_height() + voff), fontsize=14)

# now annotate the column data headings... 
ax.set_title('Engineering Change Management Documents', fontsize=18)
ax.set_ylabel('Number of MOC Documents', fontsize=16)
ax.set_xlabel('Project', fontsize=18)
ax.legend(loc='best', fontsize=14)

plt.savefig('MOCstats.png', bbox_inches='tight')

plt.show()
```

```python
# create data
df = pd.DataFrame([['MV32', 0, 0, 147, 116, 0, 142, 0], ['MV33', 25, 2, 120, 238, 0, 144, 14], ['Sangomar', 13, 0, 25, 20, 8, 120, 119]],
                  columns=['Project', 'EA - Engineering Amendment', 'DDN - Design Deviation Notice', 'ECN - Engineering Change Notice', 'SI - Site Instruction', 'SQ - Site Query', 'TQ - Technical Query', 'VOR - Variation Order Request'])
# view data
print(df)

hoff = 0.0
voff = 3.0

import seaborn as sns
sns.set(style="darkgrid")
sns.set_palette('pastel') ### bright.

fig, ax1 = plt.subplots(1, 1, figsize=(12,8))

# plot the grouped bar chart
df.plot(x='Project',
        kind='bar',
        width=0.80,
        stacked=False,
        rot = 1,
        # title='Engineering Change Management documents by Project', 
        fontsize=16,
        # figsize=(12,8)
        ax=ax1)

for p in ax1.patches:
    ax1.annotate(str(p.get_height()).center(4), (p.get_x() + hoff, p.get_height() + voff), fontsize=14)
    
# now annotate the column data headings... 
ax1.set_title('Engineering Change Management Documents by Project', fontsize=18)
ax1.set_ylabel('Number of Engineering MOC Documents', fontsize=16)
ax1.set_xlabel('Project', fontsize=18)
ax1.legend(loc='best', fontsize=14)

plt.savefig('MOCstats.png', bbox_inches='tight')

# plt.show() ### unneccessary.
```
